*Let us see whether we can make some pretty graphs*

```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(matrix[:,0], matrix[:,1])
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(matrix[:,0], matrix[:,1], color='r')
plt.show()
```

```python
func = lambda x : { 'largeDoses': 'r', 'smallDoses': 'b', 'didntLike': 'g' }[x]
func('largeDoses')
```

```python
np.vectorize(func)
```

```python
colors = np.vectorize(func)(labels)
colors
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(matrix[:,0], matrix[:,1], c=colors)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(matrix[:,1], matrix[:,2], c=colors)
plt.show()
```
